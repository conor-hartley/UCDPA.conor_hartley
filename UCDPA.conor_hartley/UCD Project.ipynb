```python
#One hot encoding to deal with categorical varaiable ocean_proximity, drop original variable also
df = df.join(pd.get_dummies(df.ocean_proximity)).drop(['ocean_proximity'],axis = 1)
```

```python
# Heatmap of correlations to see which variables are correlated
plt.figure(figsize= (14,6))
sns.heatmap(df.corr(), annot = True , cmap = 'coolwarm')
plt.show()
```

```python
# import train test split
# sperate target and feature variables into seperate dataframes 
# split data into train set and test set
from sklearn.model_selection import train_test_split

x = df.drop(['median_house_value'],axis = 1)
y = df['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
# Take logarithim of skewed variables
x_train['total_rooms']=np.log(x_train['total_rooms'])
x_train['total_bedrooms']=np.log(x_train['total_bedrooms'])
x_train['population']=np.log(x_train['population'])
x_train['households']=np.log(x_train['households'])

x_test['total_rooms']=np.log(x_test['total_rooms'])
x_test['total_bedrooms']=np.log(x_test['total_bedrooms'])
x_test['population']=np.log(x_test['population'])
x_test['households']=np.log(x_test['households'])
```

```python
# import linear regression 
# fit linear regression model to train set
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
# model intercept 
lm.intercept_
```

```python
# Model co-efficients for each feature variable
lm.coef_
```

```python
# Using model to predict median house value of test set
y_pred = lm.predict(x_test)
print(y_pred)
```

```python
# Scatter plot of actual test set v predicted test set from random forest Linear Regression model
plt.figure(figsize = (15,8))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
pred_y_df = pd.DataFrame({'Actual Value':y_test,'Predicted value':y_pred,'Difference': y_test-y_pred})
pred_y_df
```

```python
# Calculating the root mean square error and r-squared of the mdoel 
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

```python
# importing RandomForestRegressor and fitting model to training set
from sklearn.ensemble import RandomForestRegressor

fm = RandomForestRegressor()
fm.fit(x_train,y_train)
```

```python
# Using model to predict median house value of test set
y_pred_f = fm.predict(x_test)
print(y_pred_f)
```

```python
# Scatter plot of actual test set v predicted test set from random forest model
plt.figure(figsize = (15,8))
plt.scatter(y_test,y_pred_f)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
# Calculating the root mean square error and r-squared of the mdoel 

print(np.sqrt(mean_squared_error(y_test,y_pred_f)))
print(r2_score(y_test,y_pred_f))
```
